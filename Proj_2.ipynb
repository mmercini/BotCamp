# Instalando Bibliotecas

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
```

# Requisição

```python
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))
driver.get('https://www.etf.com/etfanalytics/etf-finder')
time.sleep(7)
```

# Expandindo Tabela

```python
expand_to_100 = driver.find_element('xpath', '/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[1]/div/div[4]/button/label/span')
#expand_to_100.click()
driver.execute_script("arguments[0].click();", expand_to_100)
```

# Definindo número de páginas

```python
num_pag = driver.find_element('xpath', '//*[@id="totalPages"]')
num_pag = num_pag.text.replace("of ", "")

num_pag = int(num_pag)

num_pag
```

# Lendo a Tabela

```python
lista_tabela = []

elemento = driver.find_element('xpath', '//*[@id="finderTable"]')   ## elemento da tabela

for pagina in range(1, num_pag + 1):
    html_table = elemento.get_attribute('outerHTML')   ## pegando o html da tabela
    tabela = pd.read_html(str(html_table))[0]     ## lendo o html da tabela utilizando o pandas
    lista_tabela.append(tabela)                   ## adicionando a tabela de cada página em uma lista
    botao_avancar = driver.find_element('xpath', '//*[@id="nextPage"]')     ## encotrando o xpath do botao avancar para pegar a tabela de cada pag
    driver.execute_script("arguments[0].click();", botao_avancar)



tabela_etf = pd.concat(lista_tabela)
tabela_etf

    
```

# Coletar Dados de Performance

```python
voltar_num_pag = driver.find_element('xpath', '//*[@id="goToPage"]')

voltar_num_pag.clear()
voltar_num_pag.send_keys('1')
voltar_num_pag.send_keys(u'\ue007')
```

```python
botao_performance = driver.find_element('xpath', '//*[@id="accordion__title-4"]/div/span')
botao_performance.click()

lista_tabela = []

elemento = driver.find_element('xpath', '//*[@id="finderTable"]')   ## elemento da tabela

for pagina in range(1, num_pag + 1):
    html_table = elemento.get_attribute('outerHTML')   ## pegando o html da tabela
    tabela = pd.read_html(str(html_table))[0]     ## lendo o html da tabela utilizando o pandas
    lista_tabela.append(tabela)                   ## adicionando a tabela de cada página em uma lista
    botao_avancar = driver.find_element('xpath', '//*[@id="nextPage"]')     ## encotrando o xpath do botao avancar para pegar a tabela de cada pag
    driver.execute_script("arguments[0].click();", botao_avancar)



tabela_rentabilidade_etf = pd.concat(lista_tabela)
tabela_rentabilidade_etf
```

```python
driver.quit()
```

# Tratando Dados

```python
tabela_rentabilidade_etf = tabela_rentabilidade_etf.replace('--', pd.NA)
tabela_rentabilidade_etf = tabela_rentabilidade_etf.dropna()
```

```python
for coluna in tabela_rentabilidade_etf.columns:

    tabela_rentabilidade_etf[coluna] = tabela_rentabilidade_etf[coluna].str.rstrip('%').astype(float)/100
    
 
tabela_rentabilidade_etf
```

# Juntando as duas Tabelas

```python

dados_finais = tabela_etf.join(tabela_rentabilidade_etf, how= 'inner')

dados_finais
```

# Filtrando Fundos Alavancados

```python
dados_finais = dados_finais[~dados_finais['Segment'].str.contains('Leveraged')]
dados_finais
```

# Criando Ranking

```python
dados_finais['rank_1_ano'] = dados_finais['1 Year'].rank(ascending= False)
dados_finais['rank_3_ano'] = dados_finais['3 Years'].rank(ascending= False)
dados_finais['rank_5_ano'] = dados_finais['5 Years'].rank(ascending= False)

dados_finais['Rank Final'] = (dados_finais['rank_1_ano'] +
                              dados_finais['rank_3_ano'] +
                              dados_finais['rank_5_ano'])
dados_finais
```

```python
dados_finais = dados_finais.sort_values(by = 'Rank Final', ascending= False)
dados_finais
```
